```python
import matplotlib.pyplot as plt

# Sample data
x = migration['Date']
z = migration['totalIncoming']
w = migration['totalOutgoing']
y = migration['total']
u = migration['totalVacc']

# Resize the figure (optional)
fig, ax1 = plt.subplots(figsize=(20, 5))

# Plot the x and y values on the first y-axis
ax1.plot(x, w, label='Outgoing')
ax1.plot(x, z, label='Incoming')
ax1.plot(x, y, label='Mobility')
ax1.set_ylabel('Mobility')

# plt.xticks(x[::30], rotation=45)
ax1.text('03-01-2019', 40, '                                 Pre-Covid', rotation=0, verticalalignment='bottom', color='black')
ax1.text('03-01-2020', 40, '                                        During Covid', rotation=0, verticalalignment='bottom', color='black')
ax1.text('03-01-2021', 40, '                                                                   PostCovid', rotation=0, verticalalignment='bottom', color='black')
ax1.axhspan(0, max(y), xmin=0.35, xmax=0.41, facecolor='gray', alpha=0.4)
ax1.axhspan(0, max(y), xmin=0.41, xmax=1, facecolor='gray', alpha=0.1)
# Create a second y-axis
ax2 = ax1.twinx()
ax2.plot(x, u, label='Vaccination', color='brown')
ax2.set_ylabel('Vaccination', color='brown')

# Add vertical lines
ax1.axvline(x='03-11-2020', color='black', linestyle='--')
ax1.axvline(x='07-01-2021', color='black', linestyle='--')



# Specify the ticks you want to display on the x-axis
ax1.set_xticks(x[::30])
ax1.set_xticklabels(x[::30], rotation=45)

# Add legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2)

# Labeling the axes and title
# ax1.set_xlabel('Date')

# Display the graph
plt.savefig('mobility.pdf', bbox_inches='tight', pad_inches=0)
plt.show()
```
