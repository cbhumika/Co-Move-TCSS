```python
migration['totalOutgoing'] = migration['gurOutgoing']+migration['ghaOutgoing']+migration['farOutgoing']+migration['gurOutgoing']+migration['easChaOutgoing']+migration['gayaOutgoing']+migration['darOutgoing']+migration['meerOutgoing']+migration['karOutgoing']+migration['muzOutgoing']
```

```python
migration['totalIncoming'] = migration['gurIncoming']+migration['ghaIncoming']+migration['farIncoming']+migration['gurIncoming']+migration['easChaIncoming']+migration['gayaIncoming']+migration['darIncoming']+migration['meerIncoming']+migration['karIncoming']+migration['muzIncoming']
```

```python
p=q=r=s=0
p1=q1=r1=s1=0
for i in range(365):
  if(data3['icon'][i]=='partly-cloudy-day' or data3['icon'][i]=='wind'):
    p+=migration['totalIncoming'][i]
    p1+=migration['totalOutgoing'][i]
  elif(data3['icon'][i]=='rain'):
    q+=migration['totalIncoming'][i]
    q1+=migration['totalOutgoing'][i]
  elif(data3['icon'][i]=='clear-day'):
    r+=migration['totalIncoming'][i]
    r1+=migration['totalOutgoing'][i]
  else:
    s+=migration['totalIncoming'][i]
    s1+=migration['totalOutgoing'][i]


print(p)
print(q)
print(r)
print(s)
print(p1+p,q1+q,r1+r,s1+s)

```

```python
p2=q2=r2=s2=0
p3=q3=r3=s3=0
for i in range(365):
  if(data1['icon'][i]=='partly-cloudy-day'):
    p2+=migration['totalIncoming'][i+365]
    p3+=migration['totalOutgoing'][i+365]
  elif(data1['icon'][i]=='rain'):
    q2+=migration['totalIncoming'][i+365]
    q3+=migration['totalOutgoing'][i+365]
  elif(data1['icon'][i]=='clear-day'):
    r2+=migration['totalIncoming'][i+365]
    r3+=migration['totalOutgoing'][i+365]
  else:
    s2+=migration['totalIncoming'][i+365]
    s3+=migration['totalOutgoing'][i+365]


print(p)
print(q)
print(r)
print(s)
print(p2+p3,q2+q3,r2+r3,s2+s3)

```

```python
weather_type = ['partly-cloudy-day', 'rain', 'clear-day', 'fog']
Before_covid = [2003,1253,1452,931]
During_covid = [2544,1593,1840,920]
After_covid = [919,614, 525,470]

x = np.arange(4)
width = 0.2
  
# plot data in grouped manner of bar type
plt.bar(x-0.2, Before_covid, width, color='tab:blue',hatch='//')
plt.bar(x, During_covid, width, color='tab:orange',hatch='o')
plt.bar(x+0.2, After_covid, width, color='tab:green',hatch='*')
plt.xticks(x, weather_type)
plt.xlabel("Weather_type",)
plt.ylabel("Mobility")
plt.legend(["Pre-covid", "Covid","Post-covid"],loc='upper right')
plt.savefig("weather vs mobility.png",bbox_inches ="tight")
f = plt.figure()
f.set_figwidth(5)
f.set_figheight(5)
plt.show(f)

# plt.bar(weather_type, Incoming_migration)
# plt.title('title name')
# plt.xlabel('weather_type')
# plt.ylabel('Incoming_migration')
# plt.show()
```
