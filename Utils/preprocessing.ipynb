```python
from datetime import date
# Function to find number of days between 2 dates
def numOfDays(date1, date2):
    return (date2-date1).days
```

```python
# day[0] represents count of users tweeting on weekend and day[1] 
# reprsents count of users tweeting on weekdays
day = [0,0]
for author_id, lst in Dict.items():
    curr_bool = [False,False]
    for i in lst:
        date1 = date(2020, 3, 1)
        date2 = date(int(i[0][:4]),int(i[0][5:7]),int(i[0][8:10]))
        diff = numOfDays(date1 , date2)
        if((diff%7 == 0 or diff%7 == 6) and curr_bool[0] == False):
            day[0] += 1
            curr_bool[0] = True
        elif((diff%7 != 0 and diff%7 != 6) and curr_bool[1] == False):
            day[1] += 1
            curr_bool[1] = True
```

```python
# calculate incoming and outgoing migrations 
for author_id, mig in Dict.items():
    # initial location of the author
    state = mig[0][1]
    for i in range(1,len(mig)): 
        if(mig[i][1] != state): # if there is change in location of tweeting , migration detected !
            date1 = date(2020, 3, 1)
            date2 = date(int(mig[i-1][0][:4]),int(mig[i-1][0][5:7]),int(mig[i-1][0][8:10]))
            date3 = date(int(mig[i][0][:4]),int(mig[i][0][5:7]),int(mig[i][0][8:10]))
            days2 = numOfDays(date1 , date2)
            days3 = numOfDays(date1 , date3)
            if(state == 'Delhi'): # for prev location delhi , outgoing migration detected !
                add = 365*code[mig[i][1]]
                final_dataset['Outgoing'][days3 + add] += 1
            elif(mig[i][1] == 'Delhi'): # else incoming migration
                add = 365*code[state]
                final_dataset['Incoming'][days3 + add] += 1
            state = mig[i][1]
```
